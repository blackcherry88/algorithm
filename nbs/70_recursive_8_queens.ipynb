```python

def eight_queens(n: int):
    if n <= 0:
        return
    
    r = [0] * n
    
    def _helper(row: int):
        if row >= n:
            yield r
            return
        
        for c in range(n):
            r[row] = c
            if _conflict(row):
                continue
            yield from _helper(row+1)
    
    def _conflict(row: int) -> bool:
        for i in range(row):
            if (r[i] == r[row]) or \
            (i + r[i] == row + r[row]) or \
            (r[i] -i == r[row] - row):
                return True
        return False
    
    yield from _helper(0)

```

```python
cnt = 0
n = 8
for s in eight_queens(n):
    cnt += 1
    print(f"{cnt}: {s}")
```
